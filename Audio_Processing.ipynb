```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the audio file
audio_file = '/content/guitar_stroke_donw.wav'
audio_data, sample_rate = librosa.load(audio_file)

# Plot the time domain
plt.figure(figsize=(10, 4))
plt.plot(audio_data)
plt.title('Time Domain of the Audio Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()
```

```python

fft_result = np.fft.fft(audio_data)
frequencies = np.fft.fftfreq(len(fft_result), 1/sample_rate)

# Plot the frequency domain
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_result)[:len(fft_result)//2])
plt.title('Frequency Domain of the Audio Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```
